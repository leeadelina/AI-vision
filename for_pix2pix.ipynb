```python


import PIL
import requests
import torch
from diffusers import StableDiffusionInstructPix2PixPipeline
model_id ="/content/logsss"
#model_id = '/content/logsss/checkpoint-20'

pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda")
generator = torch.Generator("cuda").manual_seed(42)


from PIL import Image, ImageOps

def load_image(file_path):
    image = Image.open(file_path)
    image = ImageOps.exif_transpose(image)
    image = image.convert("RGB")
    return image
def process_image(file_path):
    # 이미지 열기
    image = Image.open(file_path)
    image = ImageOps.exif_transpose(image)  # 방향 정보 보정
    image = image.convert("RGB")            # RGB로 변환

    # 이미지의 중앙을 기준으로 1:1 비율로 자르기
    width, height = image.size
    min_dim = min(width, height)
    left = (width - min_dim) / 2
    top = (height - min_dim) / 2
    right = (width + min_dim) / 2
    bottom = (height + min_dim) / 2
    image = image.crop((left, top, right, bottom))

    # 512x512 크기로 리사이즈
    image = image.resize((512, 512))

    return image
def process_image_with_padding(file_path):
    # 이미지 열기
    image = Image.open(file_path)
    image = ImageOps.exif_transpose(image)  # 방향 정보 보정
    image = image.convert("RGB")            # RGB로 변환

    # 1:1 비율로 여백 추가
    image = ImageOps.pad(image, (512, 512), color=(255, 255, 255))  # 흰색 배경 추가

    return image
def download_image(url):
   image = PIL.Image.open(requests.get(url, stream=True).raw)
   image = PIL.ImageOps.exif_transpose(image)
   image = image.convert("RGB")
   return image
```

```python
import os
from PIL import Image, ImageOps
import torch
from diffusers import StableDiffusionInstructPix2PixPipeline
import matplotlib.pyplot as plt
# 이미지 처리 함수들
def process_image(file_path):
    image = Image.open(file_path)
    image = ImageOps.exif_transpose(image).convert("RGB")
    min_dim = min(image.size)
    image = image.crop(((image.width - min_dim) / 2, (image.height - min_dim) / 2, (image.width + min_dim) / 2, (image.height + min_dim) / 2))
    return image.resize((512, 512))

def process_image_with_padding(file_path):
    image = Image.open(file_path)
    image = ImageOps.exif_transpose(image).convert("RGB")
    return ImageOps.pad(image, (512, 512), color=(255, 255, 255))

# 이미지 생성 함수
def generate_and_display_images(file_path, prompts, output_dir="/content/generated_images"):
    os.makedirs(output_dir, exist_ok=True)
    image_processing_functions = [process_image, process_image_with_padding]

    for func in image_processing_functions:
        processed_image = func(file_path)
        
        for prompt in prompts:
            # 생성된 이미지 생성
            edited_image = pipe(
                prompt=prompt,
                image=processed_image,
                num_inference_steps=100,
                image_guidance_scale=1.5,
                guidance_scale=8,
                generator=generator
            ).images[0]
            
            # 파일 저장
            file_name = os.path.basename(file_path).split('.')[0]
            func_name = func.__name__
            prompt_short = "_".join(prompt.split()[:3])  # 첫 3개의 단어로 프롬프트 요약
            save_path = os.path.join(output_dir, f"{file_name}_{func_name}_{prompt_short}.png")
            edited_image.save(save_path)
            print(f"Saved {save_path}")

            # 원본 이미지, 프롬프트, 생성된 이미지 출력
            display_images(processed_image, prompt, edited_image)

# 이미지 비교 출력 함수
def display_images(original, prompt, edited):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle(f"Prompt: {prompt}", fontsize=16)

    # 원본 이미지
    axs[0].imshow(original)
    axs[0].axis("off")
    axs[0].set_title("Original Image")

    # 프롬프트 텍스트
    axs[1].text(0.5, 0.5, prompt, ha="center", va="center", fontsize=12)
    axs[1].axis("off")
    axs[1].set_title("Prompt")

    # 생성된 이미지
    axs[2].imshow(edited)
    axs[2].axis("off")
    axs[2].set_title("Edited Image")

    plt.show()

# 각 파일과 프롬프트 리스트
file_prompts = {
    "ftp_drive/samples/test1.png": ["change Haegeum Fiddle to Violin"],
    "ftp_drive/samples/test2.png": ["change Dongnaehakchum to Ballet"],
    "ftp_drive/samples/go000015_KimJungeun.jpg": [
        "Korean instrument Janggu Drum change to Timpani",
        "Janggu Drum change to Timpani",
        "Janggu change to Timpani",
        "change Yanggeum Zither to Zither",
        "change Yanggeumto Zither",
        "Yanggeum to Zither"
    ],
    "ftp_drive/samples/go000006.jpg": [
        "Korean instrument Janggu Drum change to Timpani",
        "Buk change to drum",
        "change Haegeum Fiddle to Violin",
        "change Haegeum to Violin",
        "Haegeum to Violin"
    ],
    "ftp_drive/samples/go000017_KinJungeun.jpg": ["Yanggeum to Zither"]
}

# 모든 파일에 대해 이미지 생성 실행
for file_path, prompts in file_prompts.items():
    generate_and_display_images(file_path, prompts)
```
